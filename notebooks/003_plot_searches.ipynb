```python
import polars as pl
from matplotlib import pyplot as plt
import numpy as np
from pathlib import Path
import vizta
from loguru import logger

vizta.mpl.set_theme(context="talk")


# sage searches drop a results.sage.parquet
# ionmesh searches drop a features.csv


def extract_runtime(data_dir):
    data_dir = Path(data_dir)
    if not data_dir.is_dir():
        raise ValueError(f"{data_dir} is not a directory")

    matching = list(data_dir.glob("*.log")) + list(
        (data_dir / "../").glob(f"{data_dir.stem}.log")
    )
    if len(matching) == 0:
        raise ValueError(f"Could not find a log file in {data_dir}")

    if len(matching) > 1:
        raise ValueError(f"Found multiple log files in {data_dir}")

    # The last 3 lines should look like this:
    # real	0m41.228s
    # user	2m23.056s
    # sys	0m27.282s

    with open(matching[0], "r") as f:
        lines = f.readlines()

    if len(lines) < 3:
        raise ValueError(f"Log file {matching[0]} is too short")

    perf_section = lines[-3:]
    out = {}
    for line in perf_section:
        parts = line.split("\t")
        if len(parts) != 2:
            raise ValueError(f"Could not parse line {line}")
        time = parts[1].strip()
        split_time = time.split("m")
        if len(split_time) != 2:
            raise ValueError(f"Could not parse time {time}, parts: {split_time}")

        out[parts[0]] = int(split_time[0]) * 60 + float(split_time[1].replace("s", ""))

    return out


def read_data_dir(data_dir):
    data_dir = Path(data_dir)
    if not data_dir.is_dir():
        raise ValueError(f"{data_dir} is not a directory")

    df = None
    if (data_dir / "results.sage.parquet").exists():
        df = pl.read_parquet(data_dir / "results.sage.parquet")
    if (data_dir / "features.csv").exists():
        df = pl.read_csv(data_dir / "features.csv")

    if df is not None:
        return df
    raise ValueError(f"Could not find a valid file in {data_dir}")


def target_decoy_plot(df: pl.DataFrame, title: str):
    discriminant_col = (
        "sage_discriminant_score"
        if "sage_discriminant_score" in df.columns
        else "discriminant_score"
    )
    ms2_err_col = "fragment_ppm" if "fragment_ppm" in df.columns else "average_ppm"
    print(df.columns)

    # Keep best score per peptide
    df = df.sort(discriminant_col, descending=True).group_by("peptide").head(1)
    if "is_decoy" in df.columns:
        targets = df.filter(pl.col("is_decoy").is_not())
        decoys = df.filter(pl.col("is_decoy"))
    else:
        targets = df.filter(pl.col("label") > 0)
        decoys = df.filter(pl.col("label") < 0)

    logger.info(f"{len(targets)} targets ; {len(decoys)} decoys")

    target_discriminants = targets.select(pl.col(discriminant_col))[
        discriminant_col
    ].to_numpy()
    decoy_discriminants = decoys.select(pl.col(discriminant_col))[
        discriminant_col
    ].to_numpy()

    decoy_avg_ppm = (
        decoys.filter(pl.col("peptide_q") < 0.01)
        .select(pl.col(ms2_err_col))[ms2_err_col]
        .to_numpy()
    )
    target_avg_ppm = (
        targets.filter(pl.col("peptide_q") < 0.01)
        .select(pl.col(ms2_err_col))[ms2_err_col]
        .to_numpy()
    )

    logger.info(
        f"Targets: {len(target_avg_ppm)} ; Decoys: {len(decoy_avg_ppm)} < 0.01 FDR"
    )

    bins_discriminants = np.linspace(
        min(target_discriminants.min(), decoy_discriminants.min()),
        max(target_discriminants.max(), decoy_discriminants.max()),
        100,
    )
    bins_ppm = np.linspace(
        min(target_avg_ppm.min(), decoy_avg_ppm.min()),
        max(target_avg_ppm.max(), decoy_avg_ppm.max()),
        100,
    )
    fig, ax = plt.subplots(2)
    ax[0].hist(
        target_discriminants,
        bins=bins_discriminants,
        alpha=0.5,
        label="Targets",
    )
    ax[0].hist(
        decoy_discriminants,
        bins=bins_discriminants,
        alpha=0.5,
        label="Decoys",
    )
    ax[0].set_title("Discriminant Score")
    ax[0].legend()

    ax[1].hist(target_avg_ppm, bins=bins_ppm, alpha=0.5, label="Targets")
    ax[1].hist(decoy_avg_ppm, bins=bins_ppm, alpha=0.5, label="Decoys")
    ax[1].set_title("Average MS2 error (ppm)")
    ax[1].legend()

    fig.suptitle(title)

    timings = extract_runtime(data_dir)
    timings = {k + "_time_seconds": v for k, v in timings.items()}

    outs = {
        "fig": fig,
        "ax": ax,
        "target_discriminants": target_discriminants,
        "target_avg_ppm": target_avg_ppm,
        "decoy_discriminants": decoy_discriminants,
        "decoy_avg_ppm": decoy_avg_ppm,
        "num_targets": len(target_avg_ppm),
        "df": df.filter(pl.col("peptide_q") < 0.01),
    }
    outs.update(timings)

    return outs


def results_dict_to_df(results_dict):
    rows = []
    for title, results in results_dict.items():
        rows.append(
            {
                "experiment": title,
                "num_targets": results["num_targets"],
                "target_avg_ppm": results["target_avg_ppm"].mean(),
                "decoy_avg_ppm": results["decoy_avg_ppm"].mean(),
                "target_discriminants": results["target_discriminants"].mean(),
                "decoy_discriminants": results["decoy_discriminants"].mean(),
                "user_time_seconds": results["user_time_seconds"],
                "sys_time_seconds": results["sys_time_seconds"],
                "real_time_seconds": results["real_time_seconds"],
            }
        )
    df = pl.DataFrame(rows)
    return df


def read_and_plot(data_dir, title):
    df = read_data_dir(data_dir)
    plt_outs = target_decoy_plot(df, title)
    plt.tight_layout()
    plt.show()
    return plt_outs


HELA_EXPERIMENTS = {
    "HeLa 25ng ionmesh DIA - Native": "../outputs/PRTC_ionmesh_wide_dia",
    "HeLa 25ng sage DDA - Native": "../outputs/PRTC_sage_wide_dda",
    "HeLa 25ng sage DIA - mzML": "../outputs/PRTC_sage_wide_dia_mzml",
}

plt_outs = {}
for title, data_dir in HELA_EXPERIMENTS.items():
    plt_outs[title] = read_and_plot(data_dir, title)

bins_all_ppm = np.linspace(
    min(
        min(
            plt_outs[title]["target_avg_ppm"].min(),
            plt_outs[title]["decoy_avg_ppm"].min(),
        )
        for title in HELA_EXPERIMENTS
    ),
    max(
        max(
            plt_outs[title]["target_avg_ppm"].max(),
            plt_outs[title]["decoy_avg_ppm"].max(),
        )
        for title in HELA_EXPERIMENTS
    ),
    100,
)

# target ppm offsets among all
fig, ax = plt.subplots()
for title in HELA_EXPERIMENTS:
    if "Native" not in title:
        # I am removing it here bc it is not a 1:1 comparisson...
        # They were not acquired at the same time AND the flattener
        # Uses the dinamically linked library instead of the native
        # reader.
        continue
    ax.hist(
        plt_outs[title]["target_avg_ppm"],
        bins=bins_all_ppm,
        alpha=0.5,
        label=title.replace(" - Native", "").replace("HeLa 25ng ", ""),
    )
ax.set_title("Average MS2 error (ppm)")
ax.set_ylabel("Count")
ax.set_xlabel("Average MS2 error (ppm)")
ax.legend()
plt.show()

df = results_dict_to_df(plt_outs)
df
```

```python
# NMIAA_ionmesh_wide_dia   NMIAA_sage_wide_dda_mzml
# NMIAA_sage_wide_dda      NMIAA_sage_wide_dia_mzml
NMIAA_RESULTS_INPUTS = {
    "NMIAA E1 ionmesh DIA - Native": "../outputs/NMIAA_ionmesh_wide_dia/231121_RH30_NMIAA_E1_DIA",
    "NMIAA E6 ionmesh DIA - Native": "../outputs/NMIAA_ionmesh_wide_dia/231121_RH30_NMIAA_E6_DIA",
    "NMIAA E1 sage DDA - Native": "../outputs/NMIAA_sage_wide_dda/231121_RH30_NMIAA_E1_DDA",
    "NMIAA E6 sage DDA - Native": "../outputs/NMIAA_sage_wide_dda/231121_RH30_NMIAA_E6_DDA",
    "NMIAA E1 sage DIA - mzML": "../outputs/NMIAA_sage_wide_dia_mzml/231121_RH30_NMIAA_E1_DIA",
    "NMIAA E6 sage DIA - mzML": "../outputs/NMIAA_sage_wide_dia_mzml/231121_RH30_NMIAA_E6_DIA",
    "NMIAA E1 sage DDA - mzML": "../outputs/NMIAA_sage_wide_dda_mzml/231121_RH30_NMIAA_E1_DDA",
    "NMIAA E6 sage DDA - mzML": "../outputs/NMIAA_sage_wide_dda_mzml/231121_RH30_NMIAA_E6_DDA",
}

plt_outs = {}
for title, data_dir in NMIAA_RESULTS_INPUTS.items():
    plt_outs[title] = read_and_plot(data_dir, title)
```

```python
peps_in_dda = (
    plt_outs["NMIAA E1 sage DDA - Native"]["df"]
    .filter(pl.col("peptide").str.contains(r"\[\+7"))
    .filter(pl.col("peptide_q") < 0.01)
    .sort("peptide_q")
)

peps_common_in_ionmesh = (
    plt_outs["NMIAA E1 ionmesh DIA - Native"]["df"]
    .filter(pl.col("peptide").str.contains(r"\[\+7"))
    .filter(pl.col("peptide_q") < 0.01)
    .sort("peptide_q")
    .filter(pl.col("peptide").is_in(peps_in_dda["peptide"]))
)
peps_common_in_ionmesh
```

```python
nmiaa_peptides = pl.concat(
    [
        x["df"]
        .with_columns(experiment=pl.Series([title]))
        .select(["experiment", "peptide", "peptide_q"])
        for title, x in plt_outs.items()
    ],
    how="vertical_relaxed",
).filter(
    pl.col("peptide_q") < 0.01, pl.col("peptide").str.contains("[+71", literal=True)
)

nmiaa_peptide_counts = (
    nmiaa_peptides.sort("peptide_q")
    .group_by("experiment")
    .agg(pl.count("peptide").alias("num_nmiaa_peptides"))
    .sort("num_nmiaa_peptides")
)

nmiaa_peptide_counts
```

```python
# Bar plot of the number of NMIAA peptides identified
fig, ax = plt.subplots()
labels = [
    x.replace(" - ", "\n").replace("NMIAA E6", "Control").replace("NMIAA E1", "NMIAA")
    for x in nmiaa_peptide_counts["experiment"]
]

vals = nmiaa_peptide_counts["num_nmiaa_peptides"].to_list()
for cond in ["NMIAA", "Control"]:
    idx = [i for i, x in enumerate(labels) if cond in x]
    ax.bar(
        [labels[i].replace(cond, "") for i in idx],
        [vals[i] for i in idx],
        label=cond,
        alpha=0.5,
    )

ax.set_title("Number of NMIAA peptides identified")
# Place legend in the top left corner
ax.legend(loc="upper left")
# Rotate x labels
plt.xticks(rotation=45)
```
